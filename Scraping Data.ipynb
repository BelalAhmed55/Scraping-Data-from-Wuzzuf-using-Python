```python
from bs4 import BeautifulSoup as bs
import requests
from urllib.request import urlopen
import pandas as pd
```

```python
url = 'https://wuzzuf.net/search/jobs/?a=hpb&filters%5Broles%5D%5B0%5D=IT%2FSoftware%20Development&q=20'
```

```python
client = urlopen(url)
```

```python
html = client.read()
```

```python
html
```

```python
client.close()
```

```python
soup = bs(html,"html.parser")
```

```python
soup
```

```python
container = soup.find_all("div",{"class" : "css-1gatmva e1v1l3u10"})
```

```python
len(container)
```

```python
bs.prettify(container[0])
```

```python
f = open("/Users/belalahmed/Desktop/Wuzzufdata.csv","a", newline="")
#file = open("/Users/belalahmed/Desktop/Wuzzufdata.csv", "w", newline="")

header = "job_title,company_name,location,job_type\n"  

#f.write(header)
```

```python

for i in container:
    jtitle = i.findAll("h2", {"class": "css-m604qf"})
    job_title = jtitle[0].text.strip() if jtitle else "N/A"

    cname = i.findAll("a", {"class": "css-17s97q8"})
    company_name = cname[0].text.strip() if cname else "N/A"

    locationn = i.findAll("span", {"class": "css-5wys0k"})
    location = locationn[0].text.strip() if locationn else "N/A"
    location = location.replace(",", "|")


    jtype = i.findAll("span", {"class": "css-1ve4b75 eoyjyou0"})
    job_type = jtype[0].text.strip() if jtype else "N/A"

    # Print for debugging
    print(f"Job Title: {job_title}")
    print(f"Company Name: {company_name}")
    print(f"Location: {location}")
    print(f"Job Type: {job_type}")
    print("_" * 50)  

    f.write(f"{job_title},{company_name},{location},{job_type}\n")

f.close()
```
